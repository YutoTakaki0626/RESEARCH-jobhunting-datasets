```python
import os
import glob

import pandas as pd
import joblib

def dumpJoblib(fileName, obj):
    with open(fileName, mode="wb") as f:
        joblib.dump(obj, f, compress=3)
        
def loadJoblib(fileName):
    with open(fileName, mode="rb") as f:
        return joblib.load(f)

df = loadJoblib('data_label_fin2/09500_09999.joblib') # 変更行
texts = df['tweet_full_text'].tolist()
df['finlabel'] = 0

df.head(2)
```

```python
standard = 9500 # 変更行
start = 9500 # 変更行
end = 9999 # 変更行

for i in range(start, end+1, 1):
    if df.loc[i, 'label'] == df.loc[i, 'label2']:
        df.loc[i, 'finlabel'] = df.loc[i, 'label']
    else:
        print('-' * 30)
        print(f'□　□　□ 現在{i}個目の作業です！！\n')
        print('1回目のラベル:{}   2回目のラベル：{}'.format(df.loc[i, 'label'],df.loc[i, 'label2']))
        print(f'ツイート本文：\n {texts[i - standard]}')
        lbl = int(input('\nどのラベルに分類しますか'))
        df.loc[i, 'finlabel'] = lbl
        print('-' * 30)
    
dumpJoblib(f'data_label_fin3/09500_09999.joblib', df) # 変更行

print('\n\n\n 保存完了　\n\n\n\n\n\n\n')
```
