```python
import os
import glob

import pandas as pd
```

```python
current_dir = os.getcwd()
month_folder = os.path.join(current_dir, 'data', '*')
months = glob.glob(month_folder)
months.sort()
months
```

```python
tweets_df = []
all_files_path = []

for month in months:
    file_path = os.path.join(month, '*.csv')
    files_path = glob.glob(file_path)
    all_files_path += files_path

all_files_path.sort()
# all_files_path
```

```python
df = pd.DataFrame()

for file in all_files_path:
    tmp = pd.read_csv(file)
    df = pd.concat([df, tmp], ignore_index=True)

len(df)
```

```python
df.to_csv('./data/all_data.csv')
```

```python
df.columns
```

```python
all_tweets = df['tweet_full_text']
all_tweets = all_tweets.tolist()
```

# COUNT Something

```python
import re

def mensetsu(tweets_list):
    men_cnt = 0
    
    for tweet in tweets_list:
        if re.search(r'面接', tweet) is not None:
            men_cnt += 1
            
    return men_cnt

mensetsu(all_tweets)
```

```python
def url(tweets_list):
    url_cnt = 0

    for tweet in tweets_list:
        if re.search(r'http', tweet) is not None:
            url_cnt += 1

    return url_cnt

url(all_tweets)
```

```python
def intern(tweets_list):
    intern_cnt = 0
    
    for tweet in tweets_list:
        if re.search(r'インターン', tweet) is not None:
            intern_cnt += 1
            
    return intern_cnt

intern(all_tweets)
```
